```python
scrum_sheet['Project Name'].nunique()
```

```python
scrum_sheet['Billing Status'] = scrum_sheet['Billing Status'].apply(lambda x: 1 if x=='Billable' else 0)
```

```python
noof_projects = pd.pivot_table(scrum_sheet, values=['Project Name'],
                                 index=['Team', 'Date'], columns=None,
                   aggfunc= lambda x: len(x.unique()),
                    margins = True, margins_name='Totals', sort=True)

noof_projects
```

```python
emp_performance = pd.pivot_table(scrum_sheet, values=['Hours', 'Billing Status', 'Project Name'],
                                 index=['Team', 'Owner', 'Date'], columns=None, 
                   aggfunc={'Hours':np.sum, 'Billing Status': lambda x: 1 if np.sum(x)>=1 else 0, 'Project Name': lambda x: len(x.unique())},
                    margins = True, margins_name='Totals', sort=True)
emp_performance
```

```python
emp_performance[emp_performance['Project Name']>7]
```

```python
emp_performance = pd.merge(emp_performance, noof_projects, 'left', left_index=True, right_index=True)
emp_performance
```

```python
emp_performance = emp_performance.reset_index()
emp_performance
```

```python
emp_performance = pd.pivot_table(emp_performance, values=['Hours', 'Billing Status', 'Project Name_x', 'Project Name_y'],
                                 index=['Team', 'Owner', 'Date'], columns=None, 
                  #  aggfunc={'Hours':np.sum, 'Billing Status':np.mean, 'Project Name': lambda x: len(x.unique())},
                    # margins = True, margins_name='Totals', 
                    sort=True)
emp_performance
```

```python
emp_performance['effeciency'] = (emp_performance['Billing Status']+1)*emp_performance['Hours']*\
(emp_performance['Project Name_x']/emp_performance['Project Name_y'])
emp_performance
```

```python
emp_performance['mean_effeciency'] = emp_performance.effeciency.groupby(level=['Team', 'Owner']).sum()
```

```python
emp_performance
```

```python
emp_performance_df = emp_performance.reset_index().sort_values('mean_effeciency', ascending=True).drop(['Date'], axis=1)
```

```python
emp_performance_df.Owner[emp_performance_df['mean_effeciency']<20].unique()
```
